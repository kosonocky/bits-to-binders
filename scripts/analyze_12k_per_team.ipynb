```python
import pandas as pd

df = pd.read_csv("../data/master_data_teams.csv")

df.columns
```

```python
df = df.fillna("NA")
```

```python
# pip install statannotations
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator

# === set these ===
df = df                      # your dataframe
y_col = "percent_of_significant_seqs"                # your continuous target
categorical_columns = ["downselection", "fast_relax_used", "md_used", "diffusion_recycle", "template_or_hotspot", "linkers_added", "folding_model", "mpnn_model", "rosetta_used"]  # your categorical columns

for cat in categorical_columns:
    d = df[[cat, y_col]].dropna().copy()
    if d.empty:
        continue
    d[cat] = d[cat].astype("category")

    # order levels by median(target), descending
    order = (
        d.groupby(cat)[y_col]
         .median()
         .sort_values(ascending=False)
         .index.tolist()
    )
    d[cat] = d[cat].cat.reorder_categories(order, ordered=True)

    # plot
    fig, ax = plt.subplots(figsize=(6, 4))
    sns.violinplot(
        data=d, x=y_col, y=cat,
        orient="h", order=order,
        cut=0,            # stop at data bounds
        scale="width",    # consistent widths
        inner="quartile", # show quartiles (optional)
        ax=ax
    )

    # counts in y-tick labels
    counts = d.groupby(cat)[y_col].size().reindex(order).astype(int)
    ax.set_yticks(range(len(order)))
    ax.set_yticklabels([f"{lvl} (n={counts[lvl]})" for lvl in order])
    ax.tick_params(axis="y", length=0)
    ax.set_ylabel(cat)
    ax.set_xlabel(y_col)

    # pairwise Mann–Whitney with Bonferroni; keep only pairs with n>=2 each
    sizes = d.groupby(cat)[y_col].size()
    levels_valid = [lvl for lvl in order if sizes[lvl] >= 2]
    pairs = [(a, b) for a, b in itertools.combinations(levels_valid, 2)]

    # if too many categories, cap to most-separated pairs by median diff
    if len(pairs) > 6:
        med = d.groupby(cat)[y_col].median()
        pairs = sorted(pairs, key=lambda ab: abs(med[ab[0]] - med[ab[1]]), reverse=True)[:6]

    if pairs:
        annotator = Annotator(
            ax, pairs, data=d, x=y_col, y=cat, order=order, orient="h"
        )
        annotator.configure(
            test="Mann-Whitney",
            comparisons_correction="BH",
            text_format="star",
            pvalue_thresholds=[(1e-4, "****"), (1e-3, "***"), (1e-2, "**"), (0.05, "*"), (1, "ns")],
            show_test_name=False,
            loc="outside",       # brackets outside the violins
            hide_non_significant=False,
            fontsize=11
        )
        annotator.apply_and_annotate()

    plt.tight_layout()
    plt.savefig(f"../results/violin_{y_col}_by_{cat}.png", dpi=300, bbox_inches="tight")
    plt.show()
    plt.close(fig)
```

```python
# pip install statannotations
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statannotations.Annotator import Annotator

# --- set these ---
y_col = "percent_of_significant_seqs"   # your continuous target
save_path = "../results/violin_final_score_by_primary_method_type.png"

# explode comma-separated categories
d = df[["primary_method_type", y_col]].dropna(subset=[y_col]).copy()
d["method"] = (
    d["primary_method_type"]
    .fillna("")
    .astype(str)
    .str.split(r"\s*,\s*")
)
d = d.explode("method")
d = d[d["method"].astype(bool)]
d["method"] = d["method"].astype("category")

# (optional) keep top-k methods by count
# top_k = 12
# keep = d["method"].value_counts().nlargest(top_k).index
# d = d[d["method"].isin(keep)].copy()

# order by median of y_col (desc)
order = (
    d.groupby("method")[y_col]
     .median()
     .sort_values(ascending=False)
     .index.tolist()
)
d["method"] = d["method"].cat.reorder_categories(order, ordered=True)

# plot
fig, ax = plt.subplots(figsize=(7, max(4, 0.35*len(order))))
sns.violinplot(
    data=d, x=y_col, y="method",
    orient="h", order=order,
    cut=0, scale="width", inner="quartile", ax=ax
)

# counts
counts = d.groupby("method")[y_col].size().reindex(order).astype(int)
ax.set_yticks(range(len(order)))
ax.set_yticklabels([f"{lvl} (n={counts[lvl]})" for lvl in order], fontsize=9)
ax.tick_params(axis="y", length=0)
ax.set_ylabel("")
ax.set_xlabel(y_col)

# pairwise tests with BH (FDR) correction
sizes = d.groupby("method")[y_col].size()
levels_valid = [lvl for lvl in order if sizes[lvl] >= 2]
pairs = list(itertools.combinations(levels_valid, 2))

# cap number of brackets if many categories
if len(pairs) > 10:
    med = d.groupby("method")[y_col].median()
    pairs = sorted(pairs, key=lambda ab: abs(med[ab[0]] - med[ab[1]]), reverse=True)[:10]

if pairs:
    annotator = Annotator(ax, pairs, data=d, x=y_col, y="method", order=order, orient="h")
    annotator.configure(
        test="Mann-Whitney",
        comparisons_correction="BH",           # <-- Benjamini–Hochberg FDR
        text_format="star",
        pvalue_thresholds=[(1e-4, "****"), (1e-3, "***"), (1e-2, "**"), (0.05, "*"), (1, "ns")],
        show_test_name=False,
        loc="outside",
        hide_non_significant=False,
        fontsize=10
    )
    annotator.apply_and_annotate()

plt.tight_layout()
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()
plt.close(fig)
```
