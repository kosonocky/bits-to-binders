```python
# build a fixed copy
fix_df = main_df[cols].copy()

ab_cols  = [c for c in fix_df.columns if c.startswith("A_B_")]
ac_cols  = [c for c in fix_df.columns if c.startswith("A_C_")]
abc_cols = [c for c in fix_df.columns if c.startswith("A_BC_")]

def _na_or_zero(s):
    return s.isna() | s.eq(0)

# collect suffixes (e.g., "_dG_cross")
sfx = set([c[3:] for c in ab_cols]) | set([c[3:] for c in ac_cols]) | set([c[4:] for c in abc_cols])

filled_ab = filled_ac = filled_abc = 0

for suf in sorted(sfx):
    ab = f"A_B{suf}"
    ac = f"A_C{suf}"
    abc = f"A_BC{suf}"

    # --- fill A_B from A_C where A_B is 0/NaN and A_C exists ---
    if ab in fix_df.columns and ac in fix_df.columns:
        mask = _na_or_zero(fix_df[ab]) & (~_na_or_zero(fix_df[ac]))
        if mask.any():
            fix_df.loc[mask, ab] = fix_df.loc[mask, ac]
            filled_ab += int(mask.sum())

    # --- fill A_C from A_B where A_C is 0/NaN and A_B exists ---
    if ac in fix_df.columns and ab in fix_df.columns:
        mask = _na_or_zero(fix_df[ac]) & (~_na_or_zero(fix_df[ab]))
        if mask.any():
            fix_df.loc[mask, ac] = fix_df.loc[mask, ab]
            filled_ac += int(mask.sum())

    # --- fill A_BC from A_B (preferred) else A_C where A_BC is 0/NaN ---
    if abc in fix_df.columns:
        # candidate value: prefer A_B if present else A_C
        cand = None
        if ab in fix_df.columns and ac in fix_df.columns:
            cand = fix_df[ab].where(~_na_or_zero(fix_df[ab]), fix_df[ac])
        elif ab in fix_df.columns:
            cand = fix_df[ab]
        elif ac in fix_df.columns:
            cand = fix_df[ac]

        if cand is not None:
            mask = _na_or_zero(fix_df[abc]) & (~_na_or_zero(cand))
            if mask.any():
                fix_df.loc[mask, abc] = cand[mask]
                filled_abc += int(mask.sum())

print(f"Filled A_B cells: {filled_ab}, A_C cells: {filled_ac}, A_BC cells: {filled_abc}")

# write out merged file (unchanged from your code)
df = pd.read_csv("../data/master_data_top10.csv")
df = df.iloc[2:]  # skip first two rows as these are controls
df["global_id"] = df["global_id"].astype(int)
fix_df.merge(df, on="global_id", how="inner").to_csv(
    "../data/master_data_top10_with_features_fixed.csv", index=False
)
```
