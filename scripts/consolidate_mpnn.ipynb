```python
import os
import pandas as pd
import torch

dirnames = [
    "autoregwithoutseq_proteinmpnn",
    "singleaascorewithoutseq_proteinmpnn",
    "autoregwithoutseq_solublempnn",
    "singleaascorewithoutseq_solublempnn",
    "autoregwithseq_proteinmpnn",
    "singleaascorewithseq_proteinmpnn",
    "autoregwithseq_solublempnn",
    "singleaascorewithseq_solublempnn",
]

df = pd.DataFrame()

aas = ["A", "C", "D", "E", "F", "G", "H", "I", "K", "L", "M", "N", "P", "Q", "R", "S", "T", "V", "W", "Y"]

for dirname in dirnames:
    print(dirname)
    dpath = f"../../LigandMPNN/outputs/bits_to_binders/{dirname}"
    sub_df = pd.DataFrame()
    for fname in os.listdir(dpath):
        if not fname.endswith(".pt"):
            continue
        fpath = os.path.join(dpath, fname)
        data = torch.load(fpath, weights_only=False)
        binder_sequence = data["native_sequence"][0:80]
        total_score = 0
        for i, aa_idx in enumerate(binder_sequence):
            try:
                total_score += data["mean_of_probs"][f"A{i+1}"][aas[aa_idx]]
            except KeyError:
                debug = data["mean_of_probs"]
                print(fname)
                raise
        total_score /= len(binder_sequence)
        sub_df = pd.concat([sub_df, pd.DataFrame({
            "dirname": [dirname],
            "filename": [fname],
            "total_score": [total_score],
        })], ignore_index=True)
    df = pd.concat([df, sub_df], ignore_index=True)
```
